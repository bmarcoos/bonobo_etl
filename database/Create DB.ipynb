```python
import sqlalchemy as db
import sys

sys.path.append("../conf")
from db_conn import engine
```

```python
metadata = db.MetaData()
```

## Yahoo Prices 

```python
table = db.Table("yahoo_prices", metadata,
                 db.Column('date', db.DATE(), primary_key=True),
                 db.Column('stock_id', db.VARCHAR(10), primary_key=True),
                 db.Column('close', db.Float(), default=None),
                 db.Column('volume', db.Float(), default=None)
)

metadata.create_all(engine)
```

## OCDE

```python
table = db.Table("ocde", metadata,
                 db.Column('variable', db.VARCHAR(50), primary_key=True),
                 db.Column('period', db.VARCHAR(20), primary_key=True),
                 db.Column('period_type', db.VARCHAR(20), default=None),
                 db.Column('year', db.Integer(), default=None),
                 db.Column('quarter', db.Integer(), default=None),
                 db.Column('month', db.Integer(), default=None),
                 db.Column('value', db.Float(), default=None),
)

metadata.create_all(engine)
```

## CMF

```python
table = db.Table("cmf", metadata,
                 db.Column('rut', db.Integer(), primary_key=True),
                 db.Column('fecha', db.DATE(), primary_key=True),
                 db.Column('moneda', db.VARCHAR(20), default=None),
                 db.Column('fecha_inicio', db.DATE(), default=None),
                 db.Column('fecha_termino', db.DATE(), default=None),
                 db.Column('ingreso', db.BIGINT(), default=None),
                 db.Column('ganancia', db.BIGINT(), default=None)

)

metadata.create_all(engine)
```

## Twitter

```python
table = db.Table("twitter", metadata,
                 db.Column('id', db.BIGINT(), primary_key=True),
                 db.Column('created_at', db.DATETIME(), primary_key=True),
                 db.Column('full_text', db.VARCHAR(150), default=None),
                 db.Column('url', db.VARCHAR(100), default=None),
                 db.Column('hashtags', db.JSON(), default=None)
)

metadata.create_all(engine)
```
